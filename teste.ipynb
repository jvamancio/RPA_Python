```python
import pandas as pd
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium .webdriver.support.ui import WebDriverWait
from selenium .webdriver.support import expected_conditions as EC 
from selenium.common.exceptions import NoSuchElementException


df=pd.read_excel("teste gaus (2).xlsx")


servico = Service(ChromeDriverManager().install())
navegador = webdriver.Chrome(service=servico)
navegador.get("https://gaus.vivo.com.br/login")

navegador.find_element(By.XPATH, '//*[@id="login"]').send_keys("")
navegador.find_element(By.XPATH, '//*[@id="password"]').send_keys("")
navegador.find_element(By.XPATH, '//*[@id="btnEntrar"]').click()

wait = WebDriverWait(navegador, 25)

for index, row in df.iterrows():
    lista = row["GAUS"]

    while True:
        try:
            menu_icon = wait.until(EC.visibility_of_element_located((By.XPATH, '//*[@id="menuPrincipal"]')))
            menu_icon.click()

            acess_icon = wait.until(EC.visibility_of_element_located((By.XPATH, '//*[@id="menu"]/div[3]/div/a')))
            acess_icon.click()

            busca_icon = wait.until(EC.visibility_of_element_located((By.XPATH, '//*[@id="menu"]/div[3]/div/div/div/a[2]')))
            busca_icon.click()    

            lupa_icon = wait.until(EC.visibility_of_element_located((By.XPATH, '//*[@id="btn-toggle-busca"]')))
            lupa_icon.click()

            n_solicitacao_icon = wait.until(EC.visibility_of_element_located((By.XPATH, '//*[@id="area-busca"]/ng-form/div/div/div[2]/div[4]/vg-field/div/div/span/div/input')))
            n_solicitacao_icon.clear()
            n_solicitacao_icon.send_keys(lista)
    
            busca_solicitacao_icon = wait.until(EC.visibility_of_element_located((By.XPATH, '//*[@id="area-busca"]/ng-form/div/div/div[5]/div[4]/div/button')))
            busca_solicitacao_icon.click()

            tres_pontos_icon = wait.until(EC.visibility_of_element_located((By.XPATH, '/html/body/ui-view/div/div[1]/ui-view/div/vg-table/div/table/tbody/tr/td[11]/vg-context-menu/div/button')))
            tres_pontos_icon.click()

            aprovar_icon = wait.until(EC.visibility_of_element_located((By.XPATH, '/html/body/ui-view/div/div[1]/ui-view/div/vg-table/div/table/tbody/tr/td[11]/vg-context-menu/div/ul/li[1]/a')))
            aprovar_icon.click()

            observacao_icon = wait.until(EC.visibility_of_element_located((By.XPATH, '//*[@id="ngdialog1"]/div[2]/div/div/div/div[2]/ng-form/vg-field/div/div/span/textarea')))
            observacao_icon.click()

            sleep(4)
            digitar_observacao_icon = wait.until(EC.visibility_of_element_located((By.XPATH, '//*[@id="ngdialog1"]/div[2]/div/div/div/div[2]/ng-form/vg-field/div/div/span/textarea')))
            digitar_observacao_icon.send_keys("Aprovado.")

            aprovar_observacao_icon = wait.until(EC.visibility_of_element_located((By.XPATH, '//*[@id="sim"]')))
            aprovar_observacao_icon.click()
            wait2 = WebDriverWait(navegador, 2)
           
            sleep(2)
            fechar_popUp_icon = wait.until(EC.visibility_of_element_located((By.XPATH, '//*[@id="toast-container"]/div/div[1]/button')))
            fechar_popUp_icon.click()

            break
        except NoSuchElementException:
            print("não encontrado")
        continue    
```
